# Analyzing H-1B visas from 2015-2019

In this project, we will analyze and find trends about where the applicants are hired, what field they are hired in, and where the demand lies for skilled foreign workers. 

H-1B visas are given to skilled foreign workers to grant them the opportunity to work within the USA for a specified amount of time. Furthermore, companies are also able to sponsor these foreign workers as long as certain educational and work experience criterias are met.

The dataset for this project originates from the U.S. Department of Labor Site and provided through
<a href="https://www.kaggle.com/abrambeyer/h1b-visa-petitions-20152019">Kaggle</a>. This dataset includes 857216 entries of aggregated data from 2015-2019 H-1B applications including but not limited to destination, titles held, companies, and case status throughout the United States.

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('H-1B 2015_2019.csv', encoding='latin-1')
```

```python
# Checking to see the count for each case status throughout 2015-2019
sns.set_style('darkgrid')
sns.set(rc={'figure.figsize':(10,7)})
sns.countplot(df['CASE_STATUS'], data = df)
plt.title('Case Status From 2015-2019')

plt.show()
```

 - Certified cases make up a large majority of all H-1B visas distributed.

```python
# Checking to see the count for each case status throughout 2015-2019 with an emphasis on full/part time positions.
sns.set(rc={'figure.figsize':(10,7)})
sns.countplot(df['CASE_STATUS'], hue = 'FULL_TIME_POSITION',data = df)
plt.title('Case Status From 2015-2019')

plt.show()
```

```python
plt.figure(figsize = (10,10))
df['JOB_TITLE'].value_counts()[:25].plot(kind='barh').invert_yaxis()
plt.xlabel('Frequency')
plt.ylabel('Profession')
plt.title('Job Frequency from 2015 - 2019')

plt.show()
```

- The majority of the top jobs that H-1B workers were recruited for were within the technology field. 

```python
plt.figure(figsize = (10,8))
df['WORKSITE_STATE_FULL'].value_counts()[:20].plot(kind='barh').invert_yaxis()
plt.xlabel('Frequency')
plt.ylabel('State')
plt.title('Job Frequency from 2015 - 2019 Within Respective States')

plt.show()
```

- New York and Chicago have the highest demand for H-1B workers. 
- This could be due to the fact that New York and Chicago are the first and third largest cities in the USA.

```python
fig = plt.figure()
ax1 = plt.subplot(2,2,1)
fig.tight_layout() 

ax1 = sns.countplot(x = 'YEAR', data = df[df['CASE_STATUS']=='CERTIFIED'])
ax1.set_title('Certified Applications')

ax2 = plt.subplot(2,2,2)
ax2 = sns.countplot(x = 'YEAR', data = df[df['CASE_STATUS']=='CERTIFIED-WITHDRAWN'])
ax2.set_title('Certified-Withdrawn Applications')

ax3 = plt.subplot(2,2,3)
ax3 = sns.countplot(x = 'YEAR', data = df[df['CASE_STATUS']=='WITHDRAWN'])
ax3.set_title('Withdrawn Applications')

ax4 = plt.subplot(2,2,4)
ax4 = sns.countplot(x = 'YEAR', data = df[df['CASE_STATUS']=='DENIED'])
ax4.set_title('Denied Applications')

plt.tight_layout()
```

- Denied applications have consistently dropped from 2105-2017 and plummeted going into 2019.
- Certified applications have continued their trend.
- Certified-Withdrawn applications have also drastically dropped in 2019
